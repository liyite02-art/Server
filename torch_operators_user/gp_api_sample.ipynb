```python
def get_factor_names():
    factor_names = list(ff.base_data_config.Base_data_dict.keys())
    factor_names = list(filter(lambda x: x not in ['label4','alpha4','label5',
                                                   'reach_up_limit','reach_down_limit',
                                                   'valid_mask_label5','valid_mask_label4_bar'],factor_names))
    return factor_names
```

```python
all_factor_names = get_factor_names()[-113:]
```

```python
# 示例1: GP训练函数, 参数见文档说明
fit_gp_formula(
    test_name = "BaseData",
    field_lst = all_factor_names,
    overwrite = True,
    # GP参数
    ic_thres = 2.0, 
    rank_ic_thres = 2.2, 
    gens = 32,
    populations = 64,
    min_depth = 2,
    max_depth = 16,
    p_crossover = 0.8,
    p_mutation = 0.2,
    init_depth = [2, 4],
)
```
